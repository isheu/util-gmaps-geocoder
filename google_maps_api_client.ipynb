```python
import urllib2
import numpy as np
import pandas as pd
from geopy import geocoders
```

```python
raw_mitre_addr = pd.read_csv('../data/final_ccn_finder.csv')
```

```python
raw_mitre_addr = raw_mitre_addr.fillna(' ')
```

```python
narrow_list = raw_mitre_addr[['ccn', 'full_address']]
```

```python
raw_mitre_addr['full_address'] = raw_mitre_addr.address_line_1 + " " + raw_mitre_addr.address_line_2 + ", " + raw_mitre_addr.city + ", " + raw_mitre_addr.state + " " + raw_mitre_addr.zipcode
raw_mitre_addr.head(30)
narrow_list = raw_mitre_addr[['mitre_id', 'full_address']]
```

```python
narrow_list.head(10)
```

```python
g = geocoders.GoogleV3()
place_list = []
lat_list = []
lng_list = []
for address in narrow_list.full_address:
    try: 
        g.geocode(address)
        place, (lat, lng) = g.geocode(address)
        place_list.append(place)
        lat_list.append(lat)
        lng_list.append(lng)
    except:
        place_list.append("Invalid Address")
        lat_list.append("Invalid Address")
        lng_list.append("Invalid Address")        
        
```

```python
place_list
```

```python
place_series = pd.Series(place_list)
lat_series = pd.Series(lat_list)
lng_series = pd.Series(lng_list)
```

```python
df_place_lat_lng = pd.DataFrame(place_series, columns = ['address'])
df_place_lat_lng['lat'] = lat_series
df_place_lat_lng['lng'] = lng_series
df_place_lat_lng['ccn'] = narrow_list['ccn']
```

```python
df_place_lat_lng.head(50)
```

```python
df_place_lat_lng.to_csv('../data/gmaps_api_output_ipps.csv')
```
